```python
# Atualizar o banco de dados, ID cliente 6840104 com redimento anual de 300000

# Conexão com o banco de dados
engine = create_engine("sqlite:///dados_clientes.db")
# engine = create_engine("mysql+pymysql://root:senha@localhost/dados_clientes")
inspector = inspect(engine)

print(inspector.get_table_names())
```

```python
# Certifique-se de que a tabela 'clientes' existe no banco de dados
dados_clientes.to_sql("clientes", engine, if_exists="replace", index=False)

query = "UPDATE clientes SET Rendimento_anual=300000 WHERE ID_Cliente=6840104"
with engine.connect() as conn:
    result = conn.execute(text(query))
    conn.commit()
# Verificando se o valor foi atualizado
query = "SELECT * FROM clientes WHERE ID_Cliente=6840104"
with engine.connect() as conn:
    result = conn.execute(text(query))
    for row in result:
        print(row)
```

```python
# Excluir o registro do cliente de ID 5008809
query = "DELETE FROM clientes WHERE ID_Cliente=5008809"
with engine.connect() as conn:
    result = conn.execute(text(query))
    conn.commit()
# Verificando se o valor foi atualizado
query = "SELECT * FROM clientes WHERE ID_Cliente=5008809"
```
